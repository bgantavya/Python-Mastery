## `if-else` Statements:
- Use `if`, `elif` (else if), and `else` to make decisions based on conditions.
- Example:
        ```python
        age = int(input("Enter your age: "))
        if age < 18:
            print("You are underage.")
        elif age == 18:
            print("Dekh kr chlaiyo beti!")
        else:
            print("Enjoy!")
        ```

2. **Nested `if-else` Statements**:
    - Handle multiple conditions within each other.
    - Example:
        ```python
        marks = int(input("Enter your marks: "))
        if marks == 33:
            print("Just pass.")
        elif marks < 33:
            if marks > 30:
                print("Can be promoted.")
            elif marks > 20:
                print("Fill compartment.")
            else:
                print("Year back.")
        else:
            if marks < 40:
                print("Need to work hard.")
            elif marks < 50:
                print("Can do better.")
            # ... (continue for other grade ranges)
        ```

3. **Greeting Based on Time**:
    - Use the `time` module to get the current hour.
    - Greet the user based on the time of day.
    - Example:
        ```python
        import time
        timestamp = time.strftime('%H')
        hour = int(timestamp)
        if hour < 12:
            print("Good morning!")
        elif hour < 16:
            print("Good afternoon!")
        elif hour < 20:
            print("Good evening!")
        else:
            print("Good night!")
        ```

4. **`match` Case Statements** (Python 3.10+):
    - A more concise way to handle multiple conditions.
    - Example:
        ```python
        x = int(input("Enter the value of 'x': "))
        match x:
            case 0:
                print("x is 0")
            case 1:
                print("x is 1")
            case _ if x != 80:
                print("Not 80")
            case _:
                print("Other value")
