# Day 1: Python Basics:
## Data Types:
  - Python supports various data types:
      - **Numbers**: Integers (`int`), floats (`float`), long integers, and complex numbers.
      - **Strings**: Enclosed in single or double quotes.
      - **Lists**, **Tuples**, and **Dictionaries**: Data structures.
      - **Boolean**: `True` or `False`.
      - **None**: Represents absence of value.

## Basic Input and Output:
   - Use `input()` to get user input (stored as a string).
   - Print output using `print()` with optional formatting (e.g., `sep`, `end`).
   For example:
Name = input("Enter your Full Name here:")
print('Hello!',Name)

## Type Casting:
   - Convert data types using functions like `int()`, `float()`, and `str()`.

## Operators:
   - **Arithmetic Operators**: `+`, `-`, `*`, `/`, `//` (floor division), `**` (exponentiation), `%` (modulo).
   - **Comparison Operators**: `<`, `>`, `<=`, `>=`, `==`, `!=`.
   - **Assignment Operators**: `=`, `+=`, `-=` (and others).
   - **Logical Operators**: `and`, `or`, `not`.
   - **Bitwise Operators**: `&`, `|`, `^`, `~`, `<<`, `>>`.
   - **Membership Operators**: `in`, `not in`.
   - **Identity Operators**: `is`, `is not`.
# Calculator: Performing different Arithmetic operations

# Input two numbers
a = float(input("Enter the first number: "))
b = float(input("Enter the second number: "))

# Addition
print(f"Addition: {a} + {b} = {a + b}")

# Subtraction
print(f"Subtraction: {a} - {b} = {a - b}")

# Multiplication
print(f"Multiplication: {a} * {b} = {a * b}")

# Exponentiation (power of)
print(f"Exponentiation: {a} ** {b} = {a ** b}")

# Division (result in float)
print(f"Division: {a} / {b} = {a / b}")

# Floor division (result in integer)
print(f"Floor Division: {a} // {b} = {a // b}")

# Modulo (remainder)
print(f"Modulo (Remainder): {a} % {b} = {a % b}")
